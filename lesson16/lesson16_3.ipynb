```python
import pandas as pd
#將DataFrame當作資料庫來操作
df = pd.read_csv('world.csv')
```

```python
df.head()  # 顯示前幾行資料
df.shape  # 顯示DataFrame的形狀
df.info()  # 顯示DataFrame的資訊
```

```python
df1 = df[['洲名','國家','日期','總確診數','新增確診數','總人口數']]
df1.head() 
df1.shape  
df1.info()  
```

```python
df1.loc[:, '日期'] = pd.to_datetime(df1['日期'])
```

```python
is_taiwan = df1['國家'] == "台灣"
df_taiwan = df1[is_taiwan]
df_taiwan.info()
```

```python
df_taiwan[df_taiwan['日期'].dt.year == 2021]
```

```python
df_taiwan[df_taiwan['日期'].dt.month == 5]
```

你已把 `日期` 轉成 datetime，篩選 2021 年資料有幾種簡單方法：

方法 1 — 用 year 屬性


```python
df_2021 = df_taiwan[df_taiwan['日期'].dt.year == 2021]
```



方法 2 — 用日期範圍


```python
df_2021 = df_taiwan[(df_taiwan['日期'] >= '2021-01-01') & (df_taiwan['日期'] < '2022-01-01')]
```



方法 3 — 把 `日期` 設為索引後用切片（方便時間序列操作）


```python
df_taiwan_idx = df_taiwan.set_index('日期')
df_2021 = df_taiwan_idx.loc['2021']
```



檢查結果：
