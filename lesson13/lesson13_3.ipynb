```python
#標準寫法
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.axes import Axes

# 1.製作figure
fig:Figure = plt.figure()

# 2.在1.的figure上製作一個axes
#實體方法
ax_1:Axes = fig.add_subplot(1, 1, 1)
ax_1.set_title('my chart')

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]


# 常用的 plot(線圖) 參數有：
# color: 線條顏色
# marker: 標記點型態
# linestyle: 線條型態
# linewidth: 線條寬度
# label: 圖例標籤
# alpha: 透明度
# markersize: 標記點大小

ax_1.plot(X, Y,color='g',marker='o',linestyle='dashed',linewidth=2,label='Data',alpha=0.8,markersize=8)
ax_1.legend()  # 顯示圖例

# 4.顯示圖
plt.show()
```
