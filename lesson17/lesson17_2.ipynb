```python
import numpy as np
import pandas as pd

df=pd.DataFrame(np.arange(16).reshape(4,4),
            index=['台北','台中','台南','高雄'],
            columns=['one','two','three','four'])
df
```

```python
#drop() 刪除列或欄
df.drop(index=['台北','高雄'], inplace=True) 
df
```

```python
#排序和排名

scores = np.random.randint(50,101,size=(50, 5))
df = pd.DataFrame(scores,
            columns=['國文','英文','數學','物理','化學'],
            index=['學生'+str(i) for i in range(1,51)])
df.index.name = '學生'
df.columns.name = '科目'
sum_values = df.sum(axis=1) 
mean_values = df.mean(axis=1) 
df['總分'] = sum_values
df['平均'] = mean_values  
df
```

```python
#排序和排名
#rank()
rank_values = sum_values.rank(method='min', ascending=False)
df['總分排名'] = rank_values
df
```

```python
#sort_values() 排序
#sort_index() 排序索引
df1 = df.sort_values(by='總分', ascending=False) #ascending=False 是由大到小排序
df1
```
