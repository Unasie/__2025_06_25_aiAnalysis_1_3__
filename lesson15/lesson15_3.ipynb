```python
#一家分店30天的營收
import numpy as np
one_store = np.random.randint(100,10000,size=(30,))
one_store
```

```python
#實體方法(累計每天的營收)
oneStoreCum = one_store.cumsum()
oneStoreCum
```

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(10, 5))
axes = figure.add_subplot(111)
axes.plot(range(1, 31), oneStoreCum, marker='o', color='r', label='Daily Revenue')
axes.set_title('Daily Revenue of One Store Over 30 Days')
axes.set_xlabel('Days')
axes.set_ylabel('Cumulative Revenue')
axes.grid(True)
axes.legend()
plt.show()
```
